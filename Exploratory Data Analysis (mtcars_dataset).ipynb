### Scatter Plot (using subplot) :

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1,xlabel='disp',ylabel='mpg')
plt.scatter(df['disp'],df['mpg'],marker='*',c='green')

plt.subplot(3,2,2,xlabel='disp',ylabel='hp')
plt.scatter(df['disp'],df['hp'])

plt.subplot(3,2,3, xlabel='wt',ylabel='mpg')
plt.scatter(df['wt'],df['mpg'])

plt.subplot(3,2,4, xlabel='wt',ylabel='disp')
plt.scatter(df['wt'],df['disp'],c='red')

plt.subplot(3,2,5, xlabel='hp',ylabel='qsec')
plt.scatter(df['hp'],df['qsec'],c='black')

plt.subplot(3,2,6, xlabel='hp',ylabel='mpg')
plt.scatter(df['hp'],df['mpg'])

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.5, 
                    hspace=0.6)
plt.show()
```

```python
# InsteadPlotting these variables separtely w.r.t a Catego. feature
#It's better to have one plot for one categ. var. which shows its relation with multiple num. var.
```

```python
# so, SCATTER PLOT with color bar method represents total 4 features
# here on: x axis= any feature, y axis= Catego. feat. , c='color bar feature', s='size of marker feature'
```

```python
plt.scatter(x='mpg',y='cyl',s='disp',c='hp',data=df,alpha=0.5)
plt.colorbar(fraction=0.09);        # s= marker size , c= color bar 

# cyl- has only 3 unique values(4,6,8)
# mpg - density of 'mpg'points keeps decresing as w.r.t 'cyl'.i.e points keep separating along x axis
# disp - size of marker'as values of disp' increses w.r.t 'cyl'
# hp - color depth shows valuesof hp. it's increasing w.r.t 'cyl'
```

```python
plt.scatter(x='hp',y='vs',s='disp',c='qsec',data=df,alpha=0.5)
plt.colorbar(fraction=0.09);   # fraction zoomin the axes

# vs- has only 2 unique values(0,1):(0- straight engine,1- V shape eng)
# hp - density and no. of 'mpg'points keeps decresing as w.r.t (vs(0)data pts > vs(1) datapts.)
# disp - size of marker'as values of disp. decreases w.r.t 'vs'
# qsec - color depth shows valuesof qsec. it's increasing w.r.t 'vs'
```

```python
sns.barplot(x='am', y='drat',hue='gear',data=df)

# am- has only 2 unique values(0,1)
#  drat- it has positive correl. with am
# gear - gear(3): being only in am(0),i.e most of '3' gear engines are manual
#        gear(4): engines with 4 gears almost equal in autom.(am-1) & manual(am=0) vehicles
#        gear(5): most of gear(5) engines are automatic(am=1) 
```
